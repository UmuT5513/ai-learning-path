## signal processing and time series analysis

```python
import pandas as pd
import numpy as np

# 10 günlük örnek veri
dates = pd.date_range(start="2023-01-01", periods=10, freq="D")
data = np.random.randint(10, 100, size=10)
df = pd.DataFrame({"price": data}, index=dates)
df
```

```python
df['price_rolling_avg'] = df['price'].rolling(5).mean()
```

```python
df
```

```python
print(df['price'][:5].mean())
print(df['price'][1:6].mean())
print(df['price'][2:7].mean())
```

* ilk 3 değere dikkat!

```python
df['price_rolling_max'] = df['price'].rolling(5).max()
df['price_rolling_min'] = df['price'].rolling(5).min()
df
```

## how much change in stock price in percentage

```python
df['pricer_percentage'] = df.price.pct_change(1)*100
```

## cumulative sum

```python
df['expanding_sum'] = df['price'].expanding().sum()
df
```

# difference from mean(5-day window)

```python
df['mean'] = df.price - df.price.rolling(5).mean()
df
```

# EMA

```python
df['ema'] = df.price.ewm(span=5).mean()
df
```
