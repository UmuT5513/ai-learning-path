# aggregation

## simple aggregations

```python
import pandas as pd

data = {'Product': ['A', 'B', 'A', 'B', 'A'],
        'Sales': [100, 200, 300, 400, 500]}
df = pd.DataFrame(data)

print(df)
```

### sum

```python
total_sales = df['Sales'].agg('sum')
total_sales
```

## multiple aggregations

```python
pd.DataFrame(df['Sales'].agg(['mean','sum','max']))
```

## groupby aggregation

```python
df.groupby('Product')['Sales'].agg(['mean','sum','max'])
```

# example

```python
import pandas as pd

data = {
    'Item': ['Cake', 'Cake', 'Bread', 'Pastry', 'Cake','Cake','Bread', 'Cake'],
    'Flavor': ['Chocolate', 'Vanilla', 'Whole Wheat', 'Strawberry', 'Chocolate','Vanilla','Whole Wheat','Strawberry'],
    'Price': [250, 220, 80, 120, 250, 172, 311, 193]
}

df = pd.DataFrame(data)
df
```

```python
df.groupby(['Item', 'Flavor'])['Price'].agg(['sum','mean'])
```
